## Exercício 1: Motivação Computacional para Uso de Heaps

### 1. Problema Computacional Real
**Escalonamento de tarefas com prioridades dinâmicas em sistemas operacionais ou servidores Web** (como tratamento de requisições prioritárias urgentes). Novas tarefas chegam a qualquer momento com prioridades variadas e a tarefa de maior prioridade deve ser sempre executada primeiro.

### 2. Justificativa Formal
- Uma heap binária máxima garante acesso ao maior elemento em tempo constante $O(1)$.
- Inserções de novas tarefas e remoções da tarefa mais prioritária ocorrem em tempo logarítmico $O(\log N)$.
- O uso de memória é ideal ($O(N)$ em um array contíguo) sem ponteiros extras.

### 3. Comparação com Outras Estruturas

| Estrutura | Inserção | Remoção do Máximo | Busca | Memória Adicional |
| :--- | :--- | :--- | :--- | :--- |
| **Lista Não Ordenada** | $O(1)$ | $O(N)$ | $O(N)$ | $O(1)$ |
| **Lista Ordenada** | $O(N)$ | $O(1)$ | $O(N)$ (ou $O(\log N)$) | $O(1)$ |
| **Heap Binária** | $O(\log N)$ | $O(\log N)$ | $O(N)$ | $O(1)$ |

- **Lista Não Ordenada**: Inserção rápida, mas a remoção exige buscar o maior elemento por toda a lista.
- **Lista Ordenada**: Remoção rápida, mas a inserção exige deslocar elementos na memória para manter a ordem.
- **Heap Binária**: Oferece o melhor balanço para o problema de priorização dinâmica, com custo logarítmico nas operações críticas.

### 4. Análise em Big O das Operações Críticas
- Obter tarefa mais prioritária (raiz): $O(1)$
- Agendar nova tarefa (Inserção): $O(\log N)$
- Concluir e retirar tarefa (Remoção da raiz): $O(\log N)$

```python
# Exercício 2: Estrutura da heap
class BinaryHeap:
    def __init__(self):
        self.heap = []
        self.comparisons = 0
        self.swaps = 0

    def parent_index(self, index):
        return (index - 1) // 2

    def left_child_index(self, index):
        return 2 * index + 1

    def right_child_index(self, index):
        return 2 * index + 2

    # Exercício 3: Inserção na heap
    def insert(self, valor):
        self.heap.append(valor)
        atual = len(self.heap) - 1
        self._sift_up(atual)

    def _sift_up(self, index):
        while index > 0:
            pai = self.parent_index(index)
            self.comparisons += 1
            if self.heap[index] > self.heap[pai]:
                self.heap[index], self.heap[pai] = self.heap[pai], self.heap[index]
                self.swaps += 1
                print(f"  [Sift-Up] Troca: {self.heap[index]} <-> {self.heap[pai]} nos indices {index} <-> {pai}")
                index = pai
            else:
                break

    # Exercício 4: Remoção do topo
    def extract_max(self):
        if len(self.heap) == 0:
            return None
        max_val = self.heap[0]
        ultimo_val = self.heap.pop()
        if len(self.heap) > 0:
            self.heap[0] = ultimo_val
            self._sift_down(0)
        return max_val

    def _sift_down(self, index):
        n = len(self.heap)
        while True:
            esquerdo = self.left_child_index(index)
            direito = self.right_child_index(index)
            maior = index
            if esquerdo < n:
                self.comparisons += 1
                if self.heap[esquerdo] > self.heap[maior]:
                    maior = esquerdo
            if derecho := direito < n:
                self.comparisons += 1
                if self.heap[direito] > self.heap[maior]:
                    maior = direito
            if maior != index:
                self.heap[index], self.heap[maior] = self.heap[maior], self.heap[index]
                self.swaps += 1
                print(f"  [Sift-Down] Troca: {self.heap[index]} <-> {self.heap[maior]} nos indices {index} <-> {maior}. Heap atual: {self.heap}")
                index = maior
            else:
                break

    # Exercício 5: Busca de elemento
    def contains(self, valor):
        for item in self.heap:
            self.comparisons += 1
            if item == valor:
                return True
        return False

    # Exercício 6: Remoção de valor
    def delete(self, valor):
        indice = -1
        for i in range(len(self.heap)):
            self.comparisons += 1
            if self.heap[i] == valor:
                indice = i
                break
        if indice == -1:
            return False
        if indice == len(self.heap) - 1:
            self.heap.pop()
            return True
        ultimo = self.heap.pop()
        self.heap[indice] = ultimo
        pai = self.parent_index(indice)
        if indice > 0:
            self.comparisons += 1
            if self.heap[indice] > self.heap[pai]:
                self._sift_up(indice)
                return True
        self._sift_down(indice)
        return True

    # Exercício 8: Construção da heap
    def build_heap(self, array):
        self.heap = list(array)
        self.comparisons = 0
        self.swaps = 0
        n = len(self.heap)
        ultimo_pai = (n // 2) - 1
        for i in range(ultimo_pai, -1, -1):
            self._sift_down(i)

# Exercício 7: Validação do vetor
def is_valid_heap(array):
    n = len(array)
    for i in range(n // 2):
        esquerdo = 2 * i + 1
        direito = 2 * i + 2
        if esquerdo < n:
            if array[i] < array[esquerdo]:
                return False
        if direito < n:
            if array[i] < array[direito]:
                return False
    return True

# Exercício 10: Maiores elementos vetor
def extract_k_largest(array, k):
    heap = BinaryHeap()
    heap.build_heap(array)
    maiores = []
    for i in range(k):
        valor = heap.extract_max()
        if valor is not None:
            maiores.append(valor)
    return maiores
```

## Exercício 5, 6, 7, 8, 9 e 10: Justificativas Teóricas

### Exercício 5: Busca Linear em Heap Binária
- **Algoritmo**: Busca linear $O(N)$ porque a heap não possui ordem total entre os irmãos (filho esquerdo e direito), apenas ordem parcial vertical ($pai \geq filhos$). Para buscas frequentes, use Árvores Balanceadas (AVL, Red-Black) ou Tabelas Hash. Portanto, não é possível realizar busca binária $O(\log N)$.

### Exercício 6: Remoção de Elemento Arbitrário
- **Algoritmo**: Localiza o índice via busca linear $O(N)$, troca com o último elemento, e aplica sift-up ou sift-down para restaurar a invariante da heap $O(\log N)$. O desempenho global é $O(N)$ devido à busca linear.

### Exercício 7: Validação Estrutural
- **Algoritmo**: Varre o array até a metade ($N//2$) e valida se todo pai é maior ou igual a seus filhos esquerdos e direitos. É $O(N)$ no pior caso.

### Exercício 8: Construção (Build-Heap)
- **Algoritmo**: Aplica sift-down de trás para frente a partir do último pai não-folha. Complexidade total é $O(N)$ porque a maioria dos nós está perto do nível mais baixo da árvore e faz pouquíssimas trocas. Aplica-se a operação sift-down de forma decrescente.

### Exercício 9: Build-Heap vs Inserção Incremental
- **Build-Heap**: $O(N)$ (ótimo se o array de entrada já estiver disponível).
- **Inserção Incremental**: $O(N \log N)$ (necessário se os dados chegam sob demanda em tempo real).

### Exercício 10: Heap Sort Parcial
- **Algoritmo**: Constrói a heap em $O(N)$ e extrai o maior elemento $k$ vezes em $O(k \log N)$. Complexidade: $O(N + k \log N)$. Muito mais rápido que ordenar todo o array se $k \ll N$.

```python
print('================================================================================')
print('EXERCÍCIO 2 E 3: TESTANDO INSERÇÃO')
print('================================================================================')
heap = BinaryHeap()
dados_insercao = [10, 20, 5, 15, 30, 25]
print(f'Inserindo: {dados_insercao}')
for valor in dados_insercao:
    heap.insert(valor)
print(f'Heap final: {heap.heap}')

print('\n================================================================================')
print('EXERCÍCIO 4: TESTANDO EXTRAÇÃO DO MÁXIMO')
print('================================================================================')
while len(heap.heap) > 0:
    max_val = heap.extract_max()
    print(f'Extraído: {max_val}, Heap: {heap.heap}')

print('\n================================================================================')
print('EXERCÍCIO 5 E 6: BUSCA E REMOÇÃO ARBITRÁRIA')
print('================================================================================')
for valor in [50, 30, 40, 10, 20]:
    heap.insert(valor)
print(f'Heap: {heap.heap}')
print(f'Contém 30? {heap.contains(30)}')
print(f'Deletando 30: {heap.delete(30)}')
print(f'Heap após delete: {heap.heap}')

print('\n================================================================================')
print('EXERCÍCIO 7 E 8: VALIDAÇÃO E CONSTRUÇÃO')
print('================================================================================')
array_desordenado = [12, 11, 13, 5, 6, 7, 1]
print(f'Array original: {array_desordenado} | É heap? {is_valid_heap(array_desordenado)}')
heap.build_heap(array_desordenado)
print(f'Após build_heap: {heap.heap} | É heap? {is_valid_heap(heap.heap)}')

print('\n================================================================================')
print('EXERCÍCIO 9 E 10: COMPARAÇÕES E HEAPSORT PARCIAL')
print('================================================================================')
h_inc = BinaryHeap()
for x in array_desordenado:
    h_inc.insert(x)
print(f'Incremental - Trocas: {h_inc.swaps}, Comp: {h_inc.comparisons}')
h_bd = BinaryHeap()
h_bd.build_heap(array_desordenado)
print(f'Build Heap - Trocas: {h_bd.swaps}, Comp: {h_bd.comparisons}')

print(f'3 Maiores de {array_desordenado}: {extract_k_largest(array_desordenado, 3)}')
```

## Exercício 11: Análise de Complexidade Empírica
A complexidade empírica é analisada gerando-se arrays aleatórios de tamanhos crescentes (10, 50, 100, 200, 500, 1000) e comparando a quantidade de trocas e comparações do método incremental contra o build_heap.

```python
import random

tamanhos = [10, 50, 100, 200, 500, 1000]
print(f"{'Tamanho':<10} | {'Incremental (Trocas/Comp)':<30} | {'Build Heap (Trocas/Comp)':<30}")
print("-" * 75)
for tam in tamanhos:
    dados_teste = [random.randint(1, 10000) for _ in range(tam)]

    # Incremental
    h_inc = BinaryHeap()
    for x in dados_teste:
        h_inc.insert(x)

    # Build-Heap
    h_bd = BinaryHeap()
    h_bd.build_heap(dados_teste)

    inc_stats = f"{h_inc.swaps}/{h_inc.comparisons}"
    build_stats = f"{h_bd.swaps}/{h_bd.comparisons}"
    print(f"{tam:<10} | {inc_stats:<30} | {build_stats:<30}")
```
